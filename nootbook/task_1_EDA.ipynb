```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path = "/content/drive/MyDrive/data set/MachineLearningRating_v3.txt"
data = pd.read_csv(path, sep = '|', low_memory = False)
data.head()
```

```python
data.isnull().sum()
```

```python
numerica_columns= ['TotalPremium', 'TotalClaims']
desc_stat = data[numerica_columns].describe()
print(desc_stat)
```

```python
data.dtypes
```

```python

# Convert 'TransactionMonth' column to datetime type
data['TransactionMonth'] = pd.to_datetime(data['TransactionMonth'])

# Convert 'Gender', 'MaritalStatus', 'Citizenship', 'Language', 'VehicleType', and other categorical columns to 'category' type
categorical_columns = ['IsVATRegistered', 'Citizenship', 'LegalType', 'Title', 'Language', 'Bank', 'AccountType',
                       'MaritalStatus', 'Gender', 'Country', 'Province', 'PostalCode', 'MainCrestaZone',
                       'SubCrestaZone', 'ItemType', 'mmcode', 'make', 'Model', 'bodytype', 'VehicleIntroDate',
                       'AlarmImmobiliser', 'TrackingDevice', 'NewVehicle', 'WrittenOff', 'Rebuilt', 'Converted',
                       'CrossBorder', 'CoverCategory', 'CoverType', 'CoverGroup', 'Section', 'Product',
                       'StatutoryClass', 'StatutoryRiskType']

for col in categorical_columns:
    data[col] = data[col].astype('category')
```

```python
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.show()
```

## Univariate Analysis

```python
data[numerica_columns].hist(bins =30, figsize=(10,6))
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Bar chart for categorical columns
for col in categorical_columns:
    data[col].value_counts().plot(kind='bar', figsize=(8, 5))
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Bivariate or Multivariate Analysis

```python
print(data['ZipCode'].unique())
```

```python
# Scatter plot for TotalPremium vs TotalClaims
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPremium', y='TotalClaims', data=data)
plt.title('Scatter plot of TotalPremium vs TotalClaims by VehicleType')
plt.show()

# Correlation matrix
correlation_matrix = data[['TotalPremium', 'TotalClaims']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

## Trends Over Geography

```python
# Group data by ZipCode to compare trends
zipcode_group = data.groupby('PostalCode').agg({'TotalPremium': 'mean', 'TotalClaims': 'mean'})
zipcode_group.plot(kind='bar', figsize=(10, 6))
plt.title('Average Premium and Claims by PostalCode')
plt.ylabel('Amount')
plt.show()
```

```python
zipcode_group = data.groupby('MainCrestaZone').agg({'TotalPremium': 'mean', 'TotalClaims': 'mean'})
zipcode_group.plot(kind='bar', figsize=(10, 6))
plt.title('Average Premium and Claims by PostalCode')
plt.ylabel('Amount')
plt.show()
```

```python
# Box plot for detecting outliers in TotalPremium and TotalClaim
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['TotalPremium', 'TotalClaims']])
plt.title('Box Plot for Outlier Detection')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Province', y='TotalPremium', data=data, ci=None)
plt.title('Average TotalPremium by Province')
plt.xticks(rotation=45)
plt.ylabel('Average Total Premium')
plt.xlabel('Province')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPremium', y='TotalClaims', data=data, hue='CoverType')
plt.title('TotalPremium vs TotalClaims by CoverType')
plt.xlabel('Total Premium')
plt.ylabel('Total Claims')
plt.show()
```
